```python
import pandas as pd
advertising = pd.read_csv("advertising.csv")
advertising.head()
```

```python
import seaborn as sns
```

```python
sns.pairplot(data = advertising, x_vars=['TV', 'Newspaper', 'Radio'], y_vars='Sales')
```

```python
sns.pairplot(data = advertising, x_vars=['','TV', 'Newspaper', 'Radio'], y_vars='Sales')
```

```python
sns.pairplot(data = advertising, x_vars=['TV', '','Newspaper', 'Radio'], y_vars='Sales')
```
